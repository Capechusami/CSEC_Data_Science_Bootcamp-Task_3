# Part 2: Titanic Dataset Analysis

In this part, we will perform a **complete data science workflow** on the Titanic dataset (`train.csv`).  
The steps include:

1. Data Exploration
2. Data Cleaning
3. Data Analysis using `groupby`
4. Filtering specific groups
5. Drawing Insights

```python
# Import pandas
import pandas as pd
```

## Step 1: Load Dataset

We will load the Titanic dataset (`train.csv`).  


```python
# Load the dataset
df = pd.read_csv("train.csv")

# Display the first 5 rows
df.head()
```

## Step 2: Data Exploration

We use:
- `.head()` → preview first rows
- `.info()` → check structure, data types, missing values
- `.describe()` → statistics for numerical columns

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Step 3: Data Cleaning

Tasks:
1. Fill missing values:
   - `Age` → median
   - `Embarked` → mode
2. Drop `Cabin` column (too many missing values)
3. Remove duplicate rows

```python
# Fill missing Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked with mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin column
df = df.drop(columns=["Cabin"])

# Remove duplicates
df = df.drop_duplicates()

# Verify cleaning
df.info()
```

## Step 4: Data Analysis

We use **`groupby`** to analyze survival rates:

- By gender
- By class
- Average age per class
- By age group

```python
# Survival rate by gender
df.groupby("Sex")["Survived"].mean()
```

```python
# Survival rate by class
df.groupby("Pclass")["Survived"].mean()
```

```python
# Average age per class
df.groupby("Pclass")["Age"].mean()
```

```python
# Create Age Groups: Child, Teen, Adult, Senior
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 60, 100],
    labels=["Child", "Teen", "Adult", "Senior"]
)

# Survival rate by age group
df.groupby("AgeGroup")["Survived"].mean()
```

## Step 5: Filtering Specific Groups

We filter:

1. Female passengers who survived
2. Children who survived
3. 1st class passengers who survived

```python
# Female survivors
df[(df["Sex"] == "female") & (df["Survived"] == 1)]
```

```python
# Children survivors (<12 years old)
df[(df["Age"] < 12) & (df["Survived"] == 1)]
```

```python
# 1st Class survivors
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```

# Step 6: Insights – Detailed Explanation

This step is where we interpret the results of our analysis to answer the key questions about Titanic passengers' survival.



## 1. Who was more likely to survive?

**Females had a much higher survival rate than males.**

**How we know:** We calculated survival rates by gender: `df.groupby("Sex")["Survived"].mean()`

**Example result:**
| Sex    | Survived |
|--------|----------|
| female | 0.74     |
| male   | 0.19     |

**Interpretation:** 74% of females survived compared to only 19% of males. This huge difference reflects the "women and children first" policy on the Titanic. Even without looking at class or age, gender alone is a strong predictor of survival.





## 2. Did class affect survival?

**Yes, 1st class passengers had the highest survival rates.**

**How we know:** Survival by passenger class was calculated: `df.groupby("Pclass")["Survived"].mean()`

**Example result:**
| Pclass | Survived |
|--------|----------|
| 1      | 0.63     |
| 2      | 0.47     |
| 3      | 0.24     |

**Interpretation:** 1st class passengers had 63% survival, 2nd class 47%, and 3rd class only 24%. This suggests access to lifeboats was easier for higher-class passengers. 3rd class passengers were farther from lifeboats and below deck, leading to lower survival.




## 3. Were children prioritized?

**Yes, children had higher survival rates than adults.**

**How we know:** We created age groups: `df["AgeGroup"] = pd.cut(df["Age"], bins=[0,12,18,60,100], labels=["Child","Teen","Adult","Senior"])` then `df.groupby("AgeGroup")["Survived"].mean()`

**Example results:**
| AgeGroup | Survived |
|----------|----------|
| Child    | 0.60     |
| Teen     | 0.40     |
| Adult    | 0.38     |
| Senior   | 0.20     |

**Interpretation:** Children (<12 years old) had a 60% survival rate, higher than adults (38%) and seniors (20%). This aligns with the "women and children first" protocol. Even among the same gender or class, children were prioritized.





## 4. Highest survival combination

**Female + 1st Class + Child**

**How we know:** `df[(df["Sex"]=="female") & (df["Pclass"]==1) & (df["AgeGroup"]=="Child")]["Survived"].mean()`

**Result:** Usually close to 100% survival.

**Interpretation:** Gender, class, and age combined to maximize survival chances:
- Female → higher probability
- 1st class → higher probability
- Child → higher probability
